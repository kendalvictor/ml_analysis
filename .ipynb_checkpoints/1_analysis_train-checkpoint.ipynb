```python
entries_per_person  = train_analisis.data.groupby('Patient_ID').agg("size").reset_index()
entries_per_person.columns = ["Patient_ID", "entries_per_person"]
entries_per_person.head()
```

```python
train_analisis.data = train_analisis.data.merge(entries_per_person, on=["Patient_ID"], how="left")
train_analisis.data.head(10)
```

```python
visit_per_camp  = train_analisis.data.groupby("Health_Camp_ID").agg('size').reset_index()
visit_per_camp.columns = ["Health_Camp_ID", "visit_per_camp"]
visit_per_camp.head(10)
```

```python
train_analisis.data = train_analisis.data.merge(visit_per_camp, on=["Health_Camp_ID"], how="left")
train_analisis.data.head(10)
```
