```python
#complemento para importar desde otro .ipynb
import nbimporter

#calculo
import numpy as np
import pandas as pd

#visualizacion
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline

#propios
from client import Analizer

healthcamp_analisis = Analizer("data/Health_Camp_Detail.csv")
healthcamp_analisis.data.head()
```

```python
healthcamp_first = Analizer("data/First_Health_Camp_Attended.csv")
healthcamp_first.data.head()
```

```python
healthcamp_third = Analizer("data/Thrid_Health_Camp_Attended.csv")
healthcamp_third.data.head()
```

```python
healthcamp_first.data['target'] = healthcamp_first.data['Health_Score'].apply(lambda x: int(x > 0))
healthcamp_first.data.head()
```

```python
healthcamp_second.data.columns = ['Patient_ID','Health_Camp_ID','Health_Score']
healthcamp_second.data.head()
```

```python
healthcamp_second.data['target'] = healthcamp_second.data['Health_Score'].apply(lambda x: int(x > 0))
healthcamp_second.data.head()
```

```python
healthcamp_third.data['target'] = healthcamp_third.data['Number_of_stall_visited'].apply(lambda x: int(x > 0))
healthcamp_third.data.head()
```

```python
healthcamp_third.data = healthcamp_third.data[['Patient_ID','Health_Camp_ID','target']]
healthcamp_second.data = healthcamp_second.data[['Patient_ID','Health_Camp_ID','target']]
healthcamp_first.data = healthcamp_first.data[['Patient_ID','Health_Camp_ID','target']]
```

```python
healthcamp_all = pd.concat([healthcamp_third.data, healthcamp_second.data, healthcamp_first.data])
healthcamp_all.head()
```
