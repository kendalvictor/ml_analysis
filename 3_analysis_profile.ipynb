```python
def count_all_counts(data, cols):
    for col in cols:
        print("-"*40)
        counts = pd.DataFrame(data[col].value_counts(normalize = True, dropna = False))
        print(counts)
        counts.sort_values(by=[col])
        plt.bar(
            [str(_) for _ in counts.index.values],
            counts[col],
            label=col
        )
        plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
        plt.show()
```

```python
count_all_counts(profile_analisis.data, ['Income','Education_Score','Age','City_Type', 'Employer_Category'])
```
