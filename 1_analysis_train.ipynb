## Analsis de veces de participacion por persona

```python
visit_per_camp  = train_analisis.data.groupby("Health_Camp_ID").agg('size').reset_index()
visit_per_camp.columns = ["Health_Camp_ID", "visit_per_camp"]
visit_per_camp.head(5)
```

```python
entries_per_person1['entries_per_person'].max(), entries_per_person1['entries_per_person'].min()
```
