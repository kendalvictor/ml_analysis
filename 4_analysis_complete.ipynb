```python
def compare_with_target(data, action='count'):
    for colum in data.columns.values:
        try:
            print(data.pivot_table(
                index=[colum], 
                columns=['target'],
                values=['ID'],
                aggfunc=action
            ))
            print("//"*50)
        except Exception as e:
            print("======== ERROR :: ", str(e))
```

```python
def colums_for_types(data):
    return {
        str(k).lower(): [col for col in list(v) if col in X_train.columns.values]
        for k, v in 
        data.columns.to_series().groupby(data.dtypes).groups.items()
    }
```

```python
from sklearn import preprocessing
month = preprocessing.LabelEncoder().fit(new_train.data['month'])
month_day = preprocessing.LabelEncoder().fit(new_train.data['month_day'])
fit_day = preprocessing.LabelEncoder().fit(new_train.data['day'])
fit_weekday = preprocessing.LabelEncoder().fit(new_train.data['weekday'])
fit_c1 = preprocessing.LabelEncoder().fit(new_train.data['Category1'])
fit_c2 = preprocessing.LabelEncoder().fit(new_train.data['Category2'])
fit_city = preprocessing.LabelEncoder().fit(new_train.data['City_Type'])
fit_em = preprocessing.LabelEncoder().fit(new_train.data['Employer_Category'])
```

```python
new_train.data['month'] = month.transform(new_train.data['month'])
new_train.data['month_day'] = month_day.transform(new_train.data['month_day'])
new_train.data['day'] = fit_day.transform(new_train.data['day'])
new_train.data['weekday'] = fit_weekday.transform(new_train.data['weekday'])
new_train.data['Category1'] = fit_c1.transform(new_train.data['Category1'])
new_train.data['Category2'] = fit_c2.transform(new_train.data['Category2'])
new_train.data['City_Type'] = fit_city.transform(new_train.data['City_Type'])
new_train.data['Employer_Category'] = fit_em.transform(new_train.data['Employer_Category'])
```

```python
new_test.data['month'] = month.transform(new_test.data['month'])
new_test.data['month_day'] = month_day.transform(new_test.data['month_day'])
new_test.data['day'] = fit_day.transform(new_test.data['day'])
new_test.data['weekday'] = fit_weekday.transform(new_test.data['weekday'])
new_test.data['Category1'] = fit_c1.transform(new_test.data['Category1'])
new_test.data['Category2'] = fit_c2.transform(new_test.data['Category2'])
new_test.data['City_Type'] = fit_city.transform(new_test.data['City_Type'])
new_test.data['Employer_Category'] = fit_em.transform(new_test.data['Employer_Category'])
```
